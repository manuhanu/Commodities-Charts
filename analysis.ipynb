```python
%matplotlib notebook
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
fedfunds = pd.read_csv('FEDFUNDS.csv')
fedfunds.set_index('DATE',inplace=True)
fedfunds.index = pd.to_datetime(fedfunds.index)

global_comm_index = pd.read_csv('PALLFNFINDEXM.csv')
global_comm_index.set_index('DATE',inplace=True)
global_comm_index.index = pd.to_datetime(global_comm_index.index)
```

```python
df = pd.merge_asof(fedfunds,global_comm_index,on='DATE')
df=df.dropna()
df=df.reset_index(drop=True)
```

```python
df
```

```python

x1 = df.index
x2 = df['FEDFUNDS']
z = df['PALLFNFINDEXM']
fig = plt.figure(figsize=(10,10))
ax = fig.gca(projection='3d')
surf = ax.plot_trisurf(x1,x2,z, cmap=plt.cm.jet, linewidth=0.2)
fig.colorbar( surf, shrink=0.5, aspect=5)
ax.view_init(50,50)
```
